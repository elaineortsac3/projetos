### Análise da base de dados - Saúde do Sono e Estilo de Vida

```python
#Leitura do arquivo CSV
df = pd.read_csv("saude_do_sono_estilo_vida.csv")
display(df.head())
```

```python
# inspeção das colunas
print("\nCOlunas originais: \n", df.columns.tolist())
```

```python
# Ajuste dos nomes das colunas
renames = {}
if "ID" in df.columns:
    renames["ID"] = "Identificador"
if 'Pressão sanguíneaaaa' in df.columns:
    renames['Pressão sanguíneaaaa'] = 'Pressão sanguínea'
if 'Ocupação' in df.columns:
    renames['Ocupação'] = 'Profissão'
if 'Categoria BMI' in df.columns:
    renames['Categoria BMI'] = 'Categoria IMC'

df = df.rename(columns=renames)
print("\nColunas renomeadas: \n", df.columns.tolist())
```

```python
# Transformaçao numerica
for col in ['Identificador', 'Idade', 'Duração do sono', 'Qualidade do sono', 'Nível de atividade física', 'Nível de estresse', 'Frequência cardíaca', 'Passos diários']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
print("\nTipos de dados:\n", df.dtypes)
```

```python
# 2. Média, Moda, Mediana de horas de sono para cada profissão
if 'Profissão' in df.columns and 'Duração do sono' in df.columns:
    Agrupar = df.groupby('Profissão')['Duração do sono']
    media_sono = Agrupar.mean()
    mediana_sono = Agrupar.median()
    moda_sono = Agrupar.apply(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)

    print("\nMédia de horas de sono por profissão:\n", media_sono)
    print("\nMediana de horas de sono por profissão:\n", mediana_sono)
    print("\nModa de horas de sono por profissão:\n", moda_sono)
else:
    print("\nColunas 'Profissão' ou 'Duração do sono' não encontradas.")
```

```python
#3. % de obesos entre Engenharia de Software
print("\n3. Percentual de obesos entre Engenharia de Software:")
if 'Profissão' in df.columns and 'Categoria IMC' in df.columns:
    eng = df[df['Profissão'].str.lower().str.contains('engenharia|software|engenheiro', na=False)]
    if len(eng) == 0:
        print("Nenhuma pessoa identificada como Engenharia de Software encontrada.")
    else:
        obesos = eng['Categoria IMC'].astype(str).str.lower().str.contains('obes', na=False)
        pct = obesos.sum() / len(eng) * 100
        print(f"{pct:.2f}% ({obesos.sum()} de {len(eng)})")
else:
    print("Necessita das colunas 'Profissão' e 'Categoria IMC'.")
```

```python
# 4. Advogar vs Representante de vendas: quem dorme menos? (média, usar isin)
print("\n4) Advogar vs Representante de vendas: quem dorme menos? (usando isin):")
if 'Profissão' in df.columns and 'Duração do sono' in df.columns:
    # Selecionar apenas as profissões relevantes usando isin
    filtro = df['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])
    subset = df[filtro]

    medias = subset.groupby('Profissão')['Duração do sono'].mean()
    print(medias)

    if 'Advogado(a)' in medias.index and 'Representante de Vendas' in medias.index:
        if medias['Advogado(a)'] < medias['Representante de Vendas']:
            print("\nResposta: Advogar está associado a menos horas de sono (média).")
        elif medias['Advogado(a)'] > medias['Representante de Vendas']:
            print("\nResposta: Representante de Vendas dorme menos em média.")
        else:
            print("Resposta: Ambos dormem a mesma média de horas.")
    else:
        print("Uma das profissões não foi encontrada no dataset.")
else:
    print("Necessita 'Profissão' e 'Duração do sono'.")
```

```python
# 5. Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono?
#  Enfermeiro(a)  vs Médico(a)

print("\n5) Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono? (usando isin):")
if 'Profissão' in df.columns and 'Duração do sono' in df.columns:
    # Seleciona apenas as profissões relevantes usando isin
    filtro = df['Profissão'].isin(['Enfermeiro(a)', 'Médico(a)'])
    subset = df[filtro]

    medias = subset.groupby('Profissão')['Duração do sono'].mean()
    print(medias)

    if 'Enfermeiro(a)' in medias.index and 'Médico(a)' in medias.index:
        if medias['Enfermeiro(a)'] < medias['Médico(a)']:
            print("Resposta: Enfermeiros dormem menos horas em média.")
        elif medias['Enfermeiro(a)'] > medias['Médico(a)']:
            print("\nResposta: Médicos dormem menos horas em média.")
        else:
            print("Resposta: Ambos dormem a mesma média de horas.")
    else:
        print("Uma das profissões não foi encontrada no dataset.")
```

```python
# 6. Faça um subconjunto com as colunas Identificador, Gênero, Idade, Pressão sanguínea e Frequência cardíaca.

print('\n6. Subconjunto de colunas:')
cols_sub = [c for c in ['Identificador', 'Gênero', 'Idade', 'Pressão sanguínea', 'Frequência cardíaca'] if c in df.columns]
print('Colunas presentes escolhidas:', cols_sub)
display(df[cols_sub].head() if len(cols_sub)>0 else 'Nenhuma das colunas solicitadas está presente.')
```

```python
# 7. Descubra qual a profissão menos frequente no conjunto. (Use value_counts)

print("\n7. Profissão menos frequente no conjunto:")
if 'Profissão' in df.columns:
    contagem = df['Profissão'].value_counts()
    frequencia_min = contagem.min()
    menos_frequente = contagem[contagem == frequencia_min].index.tolist()

    print('\nProfissão menos frequente:', menos_frequente, 'com', frequencia_min, 'ocorrências')
else:
    print("\nColuna 'Profissão' não encontrada.")
```

```python
# 8. Quem tem maior pressão sanguínea média, homens ou mulheres?
print("\n8. Quem tem maior pressão sanguínea média, homens ou mulheres?")
# Ajuste pressão sanguínea ['Sis', 'Dias'] separei para saber a média correta
df[['Sis', 'Dias']] = df['Pressão sanguínea'].str.split('/', expand=True).astype(float)

#Calculo da média
print('\nMédias gerais de pressão sanguínea:')
print('Sis média:', df['Sis'].mean())
print('Dias média:', df['Dias'].mean())

#Comparação homens e mulheres
if 'Gênero' in df.columns:
    medias_genero = df.groupby('Gênero')[['Sis', 'Dias']].mean()
    print('\nMédias por gênero:')
    display(medias_genero)

#Coluna de pressão média (sis + dias)/2
df['Pressão média'] = (df['Sis'] + df['Dias']) / 2
medias_pressao = df.groupby('Gênero')['Pressão média'].mean()
print('\nMédia de pressão sanguínea média por gênero:')
display(medias_pressao)

if medias_pressao['Homem']>medias_pressao['Mulher']:
    print(f"\nResposta: Homens têm maior pressão sanguínea média.",medias_pressao['Homem'])
elif medias_pressao['Homem']<medias_pressao['Mulher']:
    print(f"\nResposta: Mulheres têm maior pressão sanguínea média:", medias_pressao['Mulher'])
else:
    print("\nResposta: Ambos têm a mesma pressão sanguínea média.")
```

```python
# 9. É predominante entre os participantes dormir 8 horas por dia (considere usar Moda como medida)?
print("\n9. É predominante entre os participantes dormir 8 horas por dia (usando Moda)?")

if 'Duração do sono' in df.columns:
    moda_sono = df['Duração do sono'].mode().tolist()
    print("\nModa de horas de sono:", moda_sono)
    print(f'\nÉ normal dormir 8 horas por dia?', 8 in moda_sono or '8.0' in moda_sono)
else:
    print("\nColuna 'Duração do sono' não encontrada.")
```

```python
# 10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)
print("\n10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70?")
if 'Frequência cardíaca' in df.columns and 'Passos diários' in df.columns:
    media_alta = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
    media_baixa = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()
    print(f'\nMédia de passos diários para FC > 70: {media_alta:.2f}', 'FC <= 70:', f'{media_baixa:.2f}')
    if pd.notna(media_alta) and pd.notna(media_baixa):
        print('\nFC > 70 dá mais passos?', media_alta > media_baixa)
else:
    print("\nNecessita das colunas 'Frequência cardíaca' e 'Passos diários'.")
```
